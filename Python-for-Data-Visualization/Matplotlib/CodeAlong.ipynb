```python
import matplotlib.pyplot as  plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'-r')

plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'-r')

plt.subplot(2,2,2)
plt.plot(x,y,'*g')

plt.subplot(2,2,3)
plt.plot(x,y,'>y')

plt.subplot(2,2,4)
plt.plot(x,y, '.b')
```

```python
fig = plt.figure()

            # LEFT, BOTTOM, WIDTH, HEIGHT
ax1 = fig.add_axes([0.0,0.0,1,1])
ax1.set_title('LARGE PLOT')
ax1.plot(x,y)

ax2 = fig.add_axes([0.4,0.6,0.4,0.3])
ax2.set_title('SMALL PLOT')
ax2.plot(x,y)

ax3 = fig.add_axes([0.1,0.2,0.4,0.3])
ax3.set_title('SMALL PLOT')
ax3.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#plt.tight_layout()
axes[0].plot(x,y, 'bx')
axes[1].plot(x,y, 'or')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,3),dpi=90)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y,'.g')
axes[1][0].plot(x,y, 'xr')
axes[1][1].plot(x,y, 'oy')

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, label='X squared')
ax.plot(x*0.8, y, label='X cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='purple', linewidth=10, linestyle='-')
ax.plot(x+5, y, color='yellow', linewidth=10, linestyle='--')
ax.plot(x+10, y, color='green', linewidth=10, linestyle=':')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, 
        color='purple', 
        linewidth=1, 
        linestyle='-', 
        marker='o', 
        markersize=20, 
        markerfacecolor='red', 
        markeredgewidth=3, 
        markeredgecolor='blue'
       )
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.set_xlim([0,1])
ax.set_ylim([0,1])

ax.plot(x, y, 
        color='purple', 
        linewidth=1, 
        linestyle='--', 
       )
```
