```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
sns.set_style("whitegrid")
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
tips.head()
```

# Numerical plots

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(np.log(tips['total_bill']))
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.jointplot(x=tips['sex'].apply(lambda s: int(s=='Female')), y=tips['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(a=tips['total_bill'])
```

## Categorical plots

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.barplot(x='sex', y='tip', data=tips, estimator=np.std)
```

```python
sns.countplot(x=tips['sex'])
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', split=True, data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', jitter=True, data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', jitter=True, hue='sex', dodge=True, data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', dodge=True, data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='darkgreen')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## Grids

```python
plt.rcParams["patch.force_edgecolor"] = True

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head(1)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

## Regression plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':80})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```
